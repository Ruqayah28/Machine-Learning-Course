```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set()

#Load the dataset
df = pd.read_csv('drop_prediction.csv')

#5 rows of the dataset.This dataset contains students personal and academic information in order to predict whether the student dropout or not.
df.head()
```

```python

#This dataset consist of 10000 record and 19 attribute.
df.shape
```

```python
#This dataset consist of numerical and textual data, showing that there are some missing values.
df.info()
```

```python
#Missing values number for each feature.
print(df.isna().sum())
```

```python
#Check for duplicates: No duplicated rows.
df.duplicated()[df.duplicated()==True]
```

```python
#Stat info: general overview of the dataset.
df.describe(include='all')
```

```python
#Histogram: GPA Distrbution, left skewed.
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='GPA', color='purple')
plt.title("GPA Distribution")
plt.ylabel("Number of Students")
plt.xlabel("GPA")
plt.show()
```

```python
#Boxplot: GPA vs Dropout, clearly shows that students with high GPA did not dropout.
plt.figure(figsize=(5,5))
sns.boxplot(data=df, x='Dropout', y='GPA', color='blue')
plt.title('GPA vs Dropout')
plt.xlabel('Dropout')
plt.ylabel('GPA')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
#Boxplot: Attendance Rate vs Dropout, almost similar values so maybe attendance does npt affect dropout.
plt.figure(figsize=(5,5))
sns.boxplot(data=df, x='Dropout', y='Attendance_Rate', color='pink')
plt.title('Attendance Rate vs Dropout')
plt.xlabel('Dropout')
plt.ylabel('Attendance Rate')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
#Bar chart: Dropout Rate vs Department.
dropout_rate = df.groupby("Department")["Dropout"].mean().reset_index()
plt.figure(figsize=(10,5))
sns.barplot(data=dropout_rate, x="Department", y="Dropout")
plt.title("Dropout Rate by Department")
plt.xlabel("Department")
plt.ylabel("Dropout Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
dropout_counts = df["Dropout"].value_counts()
labels = ["Not Dropout", "Dropout"]
plt.figure(figsize=(6,6))
plt.pie(dropout_counts, labels=labels, autopct="%1.1f%%")
plt.title("Dropout Distribution")
plt.show()
```
